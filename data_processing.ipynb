```python
from tqdm import tqdm
```

```python
with open('./data/words.txt', 'rb') as file:
    lines = file.readlines()
```

```python
lines
```

```python
good_words = []
for word in tqdm(lines):
    word = word.decode().replace('\n', '')
    if len(word) == 5 and word.isalpha() and word.islower():
        good_words.append(word)
```

```python
len(good_words), good_words
```

```python
with open('./data/5_letters_words.txt', 'wb') as file:
    file.writelines([bytes(w+'\n', encoding='utf-8') for w in good_words])
```
